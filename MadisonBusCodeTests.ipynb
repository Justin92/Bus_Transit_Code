```python
def get_ax():
    fig, ax = plt.subplots(figsize=(8,8))
    ax.set_xlim(-8, 8)
    ax.set_ylim(-8, 8)
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    return ax
```

# How Many Stops are in Each Quadrant?

```python
print("North East", len(fri.get_stops_rect((0, 8), (0, 8))))
print("South East", len(fri.get_stops_rect((0, 8), (-8, 0))))
print("South West", len(fri.get_stops_rect((-8, 0), (-8, 0))))
print("North West", len(fri.get_stops_rect((-8, 0), (0, 8))))
```
